```python
from tkinter import  *
import pickle
root=Tk()


with open('mpg.pkl','rb')as fp:
    model=pickle.load(fp)
    fp.close()


horsepower=DoubleVar()
weight=IntVar()
displacement=DoubleVar()


def clear():
    horsepower.set(" ")
    weight.set('')
    displacement.set(' ')
    
clear()
f1=Frame(root)
l1=Label(f1,text="Horsepower :")
l1.config(bg="white",fg="grey",font="lucinda 15 bold")
l1.pack(side=LEFT,fill=X,expand=YES)

e1=Entry(f1,textvariable=horsepower)
e1.config(bg='white',fg='grey',font="lucinda 15 bold")
e1.pack(side=LEFT,fill=X,expand=YES,padx=10)
f1.pack(fill=BOTH,expand=YES,padx=20,pady=20)

f2=Frame(root)
l2=Label(f2,text="Weight :")
l2.config(bg="white",fg="grey",font="lucinda 15 bold")
l2.pack(side=LEFT,fill=X,expand=YES)

e2=Entry(f2,textvariable=weight)
e2.config(bg='white',fg='grey',font="lucinda 15 bold")
e2.pack(side=LEFT,fill=X,expand=YES,padx=10)
f2.pack(fill=BOTH,expand=YES,padx=20,pady=20)

f3=Frame(root)
l3=Label(f3,text="Displacement :")
l3.config(bg="white",fg="grey",font="lucinda 15 bold")
l3.pack(side=LEFT,fill=X,expand=YES)

e3=Entry(f3,textvariable=displacement)
e3.config(bg='white',fg='grey',font="lucinda 15 bold")
e3.pack(side=LEFT,fill=X,expand=YES,padx=10)
f3.pack(fill=BOTH,expand=YES,padx=20,pady=20)

b1=Button(root,text="Predict",command=lambda: predict())
b1.config(bg="blue",fg="red",font="lucinda 20 bold")
b1.pack(fill=X,expand=YES)

b2=Button(root,text="Exit",command=lambda :root.destroy())
b2.config(bg="red",fg="blue",font="lucinda 20 bold")
b2.pack(fill=X,expand=YES)

def predict():
    h=horsepower.get()
    w=weight.get()
    d=displacement.get()
    features=[[h,w,d]]
    clear()
    mileage=model.predict(features)[0]
    print("mileage is ",mileage)
    win=Toplevel(root)
    win.grab_set()
    text=f"""
    ----------------------
      Horsepower:{h:.2f}
      weight: {w:.2f}
      Displacement:{d:.2f}
    ----------------------
    Mileage: {mileage:.2f}
    """
    msg=Message(win,text=text)
    msg.config(bg="white",fg="grey",font="lucinda 15 bold")
    msg.pack(side=LEFT,fill=X,expand=YES)
    
    eb=Button(win,text="Exit",command=lambda :win.destroy())
    eb.config(bg="red",fg="blue",font="lucinda 20 bold")
    eb.pack(fill=X,expand=YES)
root.mainloop()
```
